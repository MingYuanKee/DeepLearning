```python
import re
import matplotlib.pyplot as plt
import numpy as np

# Replace with your actual file path
file_path = 'outdoor_train.txt'

# Read the content of the file
with open("outdoor_train.txt", 'r') as file:
    text = file.read()

# Define the regex patterns for extracting Epoch Pixel Loss
pixel_loss_pattern = r"Average PSNR\s+([\d.]+)"

# Find all matches in the text and convert them to floats
epoch_ploss = [float(match) for match in re.findall(pixel_loss_pattern, text)]

# Multiply x-axis values by 10
x_values = (np.arange(len(epoch_ploss)) * 10) + 10

plt.figure(figsize=(11, 6))
plt.plot(x_values, epoch_ploss, label='Pixel Loss')

# Annotate the y-value of each data point with larger fontsize
for i, j in zip(x_values, epoch_ploss):
    plt.text(i, j, f'{j:.2f}', ha='right', va='bottom', fontsize=10)  # Adjust fontsize as needed

plt.xlabel('Epoch')
plt.ylabel('PSNR')
plt.title('Outdoor Average PSNR per 10 Epochs for 30 epochs')
plt.legend()

# Show the plot
plt.show()
```
