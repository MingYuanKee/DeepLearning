```python
import re

# Replace with your actual file path
file_path = 'path_to_your_text_file.txt'

# Read the content of the file
with open("steve_epoch.txt", 'r') as file:
    text = file.read()

# Define the regex patterns for extracting Epoch Pixel Loss and Epoch FFT Loss
pixel_loss_pattern = r"Epoch Pixel Loss:\s+([\d.]+)"
fft_loss_pattern = r"Epoch FFT Loss:\s+([\d.]+)"

# Find all matches in the text and convert them to floats
epoch_ploss = [float(match) for match in re.findall(pixel_loss_pattern, text)]
epoch_floss = [float(match) for match in re.findall(fft_loss_pattern, text)]

# Print the results
print("Epoch Pixel Losses:", epoch_ploss)
print("Epoch FFT Losses:", epoch_floss)
```

```python
import matplotlib.pyplot as plt
import numpy as np
import os

plt.figure(figsize=(10, 6))
plt.plot(epoch_ploss, label='Pixel Loss')
plt.plot(epoch_floss, label='FFT Loss')
#plt.plot(epoch_psnr, label='PSNR')
plt.xlabel('Epoch')
plt.ylabel('Metrics')
plt.title('Training Metrics Over Epochs')
plt.legend()
file_path = os.path.join("./NTU/", "steve_outdoor_30epoch.png")
plt.savefig(file_path)

# Show the plot
plt.show()
```
