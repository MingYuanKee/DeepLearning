```python
import re
import matplotlib.pyplot as plt
import os

# Assuming the content is read from a file named "abc.txt"

file_path = './author_outdoor_testlog.txt'

# Read the file content
with open(file_path, 'r') as file:
    text_data = file.read()
    
psnr_ssim_pattern = r"PSNR_dehazing: (\d+\.\d+) ssim: (\d+\.\d+)"

# Extract PSNR and SSIM values using the same regex pattern
extracted_values = re.findall(psnr_ssim_pattern, text_data)

# Convert string tuples to float tuples
psnr_ssim_values = [(float(psnr), float(ssim)) for psnr, ssim in extracted_values]

# Splitting the PSNR and SSIM values into separate lists
psnr_author = [value[0] for value in psnr_ssim_values]
ssim_author = [value[1] for value in psnr_ssim_values]


file_path = './steve_outdoor_testlog.txt'

# Read the file content
with open(file_path, 'r') as file:
    text_data = file.read()
    
psnr_ssim_pattern = r"PSNR_dehazing: (\d+\.\d+) ssim: (\d+\.\d+)"

# Extract PSNR and SSIM values using the same regex pattern
extracted_values = re.findall(psnr_ssim_pattern, text_data)

# Convert string tuples to float tuples
psnr_ssim_values = [(float(psnr), float(ssim)) for psnr, ssim in extracted_values]

# Splitting the PSNR and SSIM values into separate lists
psnr_steve = [value[0] for value in psnr_ssim_values]
ssim_steve = [value[1] for value in psnr_ssim_values]



# Plotting the graph with PSNR and SSIM values
plt.figure(figsize=(10, 6))
plt.plot(psnr_author, label='PSNR author', linestyle='None', marker='o')
plt.plot(psnr_steve, label='PSNR ours', linestyle='None', marker='x')
#plt.plot(epoch_psnr, label='PSNR')
plt.xlabel('Iteration')
plt.ylabel('PSNR (dB)')
plt.title('PSNR compare')
plt.legend()
file_path = os.path.join("./", "psnr_compare_dot.png")
plt.savefig(file_path)

# Show the plot
plt.show()


# Plotting the graph with PSNR and SSIM values
plt.figure(figsize=(10, 6))
plt.plot(ssim_author, label='SSIM author', linestyle='None', marker='o')
plt.plot(ssim_steve, label='SSIM ours', linestyle='None', marker='x')
#plt.plot(epoch_psnr, label='PSNR')
plt.xlabel('Iteration')
plt.ylabel('SSIM')
plt.title('SSIM compare')
plt.legend()
file_path = os.path.join("./", "ssim_compare_dot.png")
plt.savefig(file_path)

# Show the plot
plt.show()



```
